```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go

df = pd.read_csv("data/Shared_data_responses_demographics.csv")

df
```

```python
df["Scenario"] = df["Scenario"].map({'Footbridge': 0, 'Loop': 1, 'Switch': 2})
df["survey.gender"] = df["survey.gender"].map({'Men': 0, 'Women': 1})
df["survey.education"] = df["survey.education"].map({'College': 0, 'No College': 1})
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split

[dfTrain, dfTest] = train_test_split(df, test_size=0.3, random_state=42)

xTrain = dfTrain[['Scenario', 'survey.age', 'survey.gender', 'survey.education', 'survey.political', 'survey.religious']].values
xTest = dfTest[['Scenario', 'survey.age', 'survey.gender', 'survey.education', 'survey.political', 'survey.religious']].values

yTrain = dfTrain['Outcome'].values
yTest = dfTest['Outcome'].values

linRegr = linear_model.LinearRegression()
linRegr.fit(xTrain, yTrain)

print("\tTrain R²: ", linRegr.score(xTrain, yTrain))
print("\tTest R²: ", linRegr.score(xTest, yTest))
```

```python
def calcCorrectness(x, y):
    a = 0
    for i in range(len(x)):
        if round(linRegr.predict(x[0].reshape(1, -1))[0]) == y[i]:
            a+=1
    print(round(a/len(x) * 100, 2), "%")
    
calcCorrectness(xTest, yTest)
calcCorrectness(xTrain, yTrain)
```
