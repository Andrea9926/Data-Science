```python
df["outcome_str"] = df["Outcome"].map({0 : "Omission", 1: "Commission"})
df["outcome_str"]
```

```python
fig = px.histogram(df, 
                   orientation="h", 
                   y="Continent", 
                   color="outcome_str", 
                   barmode="stack", 
                   facet_col="Scenario", 
                   barnorm="percent", 
                   title="Omission VS. Commission by Continent",
                   labels={"outcome_str": "Outcome"})

fig
```
